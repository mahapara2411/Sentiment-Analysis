```
Cleaned_text = []
for index,text in enumerate(review_df['Reviews']):
  print('Review %d:\n'%(index+1),text)
  Cleaned_text.append(text)
```

```
Cleaned_text
```

```
corpus = [] #Create an empty corpus
ps = PorterStemmer() # For Stemming the words to original form
stop_words = set(stopwords.words('english')) # removes words like will, i, and, it etc
for i in range(0,70):
  #removes handles (@), numbers, urls emojis and any other special charcters to have only text
  Cleaned_text_cln = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|[0-9]", ' ',Cleaned_text[i])
  # Convert all words to lower case
  Cleaned_text_cln = Cleaned_text_cln.lower()
  # Split sentences in to words
  Cleaned_text_cln = Cleaned_text_cln.split()
  # Remove english stopwords
  Cleaned_text_cln = [x for x in Cleaned_text_cln if not x in stop_words]
  # Lemmatize words to the base form
  wn = nltk.WordNetLemmatizer()
  Cleaned_text_cln = [wn.lemmatize(x) for x in Cleaned_text_cln]
  # Join words to form the original sentence, but cleaned-up
  Cleaned_text_cln = " ".join(Cleaned_text_cln)
  #Append to the list to get all the tweets in one place
  corpus.append(Cleaned_text_cln)
```

```
#Clean Corpus
corpus[0:10]
```

```
#Strip the records and create a word list for each tweet
word_list = []
for record in (review_df):
    #print(record)
    words = []
    tokens = record.split()
    #print(tokens)
    for token in tokens:
        words.append(token.lower())
    word_list.append(words)  
    
word_list
```

```
#Bag of words from the corpus
bagofwords = []
for i in range(0,70):
  words = corpus[i].split()
  bagofwords.append(words)
```

```
bagofwords[5]
```

```
#Most frequent words
all_words = list(itertools.chain(*bagofwords))
print(all_words[0:10])
print()
print('Most commonly occuring words and frequency\n')
words_freq = collections.Counter(all_words)
words_freq.most_common(20)
```

```
#Frequency of words
freq_word_df = pd.DataFrame(words_freq.most_common(15), columns=['Words','Freq'])
freq_word_df.head()
```

```
fig,ax = plt.subplots(figsize=(15,10))
freq_word_df.sort_values(by='Freq').plot.barh(x='Words', y = 'Freq', ax=ax, color = "BLUE")
plt.show()
```

```
wordfreq = {}
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

```
sentiment_text = [TextBlob(Review) for Review in corpus]
print(sentiment_text[5].polarity)
print(sentiment_text[5])
```

```
sentiments = [[Review.sentiment.polarity, str(Review)] for Review in sentiment_text]
sentiment_df = pd.DataFrame(sentiments, columns = ["Polarity", "Review_Text"])
sentiment_df.sort_values(by= 'Polarity', ascending=False)
```
